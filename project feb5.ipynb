```python
# Defining a function to separate categorical and numerical data
def separate_categorical_and_continuous(data):
    categorical_cols, numerical_cols = [] , []
    for col in data.columns:
        if data[col].dtypes == "object":
            categorical_cols.append(col)
        elif data[col].dtypes in ["int64", "float64"]:
            numerical_cols.append(col)
    return categorical_cols, numerical_cols

categorical_cols, numerical_cols = separate_categorical_and_continuous(data)
print("Categorical columns:", categorical_cols)
print("Numerical columns:", numerical_cols)
```

```python
def univariate_analysis(data):
    for col in data.columns:
        if data[col].dtypes == "object":
            plt.figure(figsize=[12, 6])
            # This is for the first subplot
            plt.subplot(1, 2, 1)
            sns.barplot(x=data[col].value_counts().index,
                        y=data[col].value_counts().values,
                        color="darkblue")
            plt.title("Categorical Distribution of {}".format(col))
            plt.subplot(1, 2, 2)
            plt.pie(x=data[col].value_counts().values,
                    labels=data[col].value_counts().index,
                    autopct='%1.2f%%')
            plt.title("Categorical Distribution of {}".format(col))
            plt.show()

        elif data[col].dtypes in ["int64", "float64"]:
            plt.figure(figsize=[12, 6])
            plt.subplot(1, 2, 1)
            sns.histplot(x=col, data=data, color="darkblue")
            plt.title("Continuous Distribution of {}".format(col))
            plt.subplot(1, 2, 2)
            sns.boxplot(x=col, data=data, color="darkblue")
            plt.title("Continuous Distribution of {}".format(col))
            plt.show()


univariate_analysis(data)
```
